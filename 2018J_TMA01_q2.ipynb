```python
import pandas as pd


#importing the rural urban classification
url = "http://geoportal1-ons.opendata.arcgis.com/datasets/402c0ed7ecaa4d838fd742de033e5e9a_0.csv"
rural_classification_df = pd.read_csv(url)

```

The first data set containing the conditional cautioning is spread in 4 different quarters. We import only the dataset included in 2010-2011 quarters as required. 
Below we import data set broken down by CPS area. The Q4 has the specified dataset broken down either for pre and post charges. 

```python
#importing the 4 quarters 2010-2011
url = "https://webarchive.nationalarchives.gov.uk/20160701142740/http://www.cps.gov.uk/data/conditional_cautioning/Q1_10-11_Conditional_Cautions_No_Administered.csv"
cc1_df = pd.read_csv(url)
url = "https://webarchive.nationalarchives.gov.uk/20160701142743/http://www.cps.gov.uk/data/conditional_cautioning/Q2_10-11_Conditional_Cautions_No_Administered.csv"
cc2_df = pd.read_csv(url)
url = "https://webarchive.nationalarchives.gov.uk/20160701142736/http://www.cps.gov.uk/data/conditional_cautioning/Q3_10-11_Conditional_Cautions_No_Administered.csv"
cc3_df = pd.read_csv(url)
url = "https://webarchive.nationalarchives.gov.uk/20160701142742/http://www.cps.gov.uk/data/conditional_cautioning/Q4_10-11_Conditional_Cautions_No_Administered_pre_charge.csv"
cc4_df = pd.read_csv(url)
url = "https://webarchive.nationalarchives.gov.uk/20160701142742/http://www.cps.gov.uk/data/conditional_cautioning/Q4_10-11_Conditional_Cautions_No_Administered_post_charge.csv"
cc4p_df = pd.read_csv(url)
```

In order to do so we rename the dataset (so that we can handle them better).

```python
#we rename and select the data needed - note that we also stripe the "Area " field from the extra space
cc1_df.rename(columns = {"Area " : "Area",'Q1 2010/2011' : 'Tot_CC'}, inplace = True)
cc1_df = cc1_df[['Area', 'Tot_CC']]

cc2_df.rename(columns={'Area ': 'Area','Q2 2010/2011': "Tot_CC"}, inplace=True)
cc2_df = cc2_df[['Area', 'Tot_CC']]

cc3_df.rename(columns={'Area ': 'Area','Q3 2010/2011': "Tot_CC"}, inplace=True)
cc3_df = cc3_df[['Area', 'Tot_CC']]

cc4_df.rename(columns={'Area ': 'Area','Q4 2010/2011': "Tot_CC"}, inplace=True)
cc4_df = cc4_df[['Area', 'Tot_CC']]

cc4p_df.rename(columns={'Area ': 'Area','Q4 2010/2011': "Tot_CC"}, inplace=True)
cc4p_df = cc4p_df[['Area', 'Tot_CC']]

#we unify and summ the total of conditional cautions per city
new_cc = pd.concat((cc1_df,cc2_df,cc3_df,cc4_df,cc4p_df)).groupby('Area',as_index=False).sum()
new_cc
```

#### 6. Put the data into an appropriate form for plotting

```python
import matplotlib.pyplot as plt

merged_4_plot_df["C.Caution % per Urban p."]=merged_4_plot_df["Tot_CC"]/merged_4_plot_df["Tot_Urban"]*100
merged_4_plot_df["C.Caution % per Rural p."]=merged_4_plot_df["Tot_CC"]/merged_4_plot_df["Tot_Rural"]*100
merged_4_plot_df["C.Caution % per Tot p."]=merged_4_plot_df["Tot_CC"]/(merged_4_plot_df["Tot_Urban"]+merged_4_plot_df["Tot_Rural"])*100
```

```python
merged_4_plot_df=merged_4_plot_df[["Area","C.Caution % per Urban p.","C.Caution % per Rural p.","C.Caution % per Tot p."]]
```

#### 7. Visualise the data

```python
merged_4_plot_df.sort_values('C.Caution % per Tot p.').set_index("Area").plot.bar(title="Percentage of Conditional Cautions give in England in Urban or Rural Area", figsize=(20,10))
plt.xlabel('Location')
plt.ylabel('Percentage %')
```
